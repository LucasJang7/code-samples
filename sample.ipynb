```python
import pandas as pd
import  matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
```

```python
# data set : https://www.kaggle.com/datasets/joebeachcapital/30000-spotify-songs/data?select=spotify_songs.csv
data = pd.read_csv("spotify_songs.csv")
data.head(5)
```

```python
sns.histplot(data, x='playlist_subgenre', hue='playlist_genre')
plt.xticks(rotation=90)
plt.show()
```

```python
data = data.drop(['track_name','track_id','playlist_id','track_album_id'], axis = 1)
```

```python
averages = data[['playlist_name','track_popularity', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']].groupby('playlist_name').mean()
```

```python
averages.hist()
plt.show()
```

```python
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(averages.corr(), annot=True, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Plot distribution of key features: Track Popularity, Danceability, and Energy
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Distribution of Track Popularity
sns.histplot(data['track_popularity'], bins=20, kde=True, ax=ax[0])
ax[0].set_title('Distribution of Track Popularity')
ax[0].set_xlabel('Track Popularity')
ax[0].set_ylabel('Frequency')

# Distribution of Danceability
sns.histplot(data['danceability'], bins=20, kde=True, ax=ax[1])
ax[1].set_title('Distribution of Danceability')
ax[1].set_xlabel('Danceability')
ax[1].set_ylabel('Frequency')

# Distribution of Energy
sns.histplot(data['energy'], bins=20, kde=True, ax=ax[2])
ax[2].set_title('Distribution of Energy')
ax[2].set_xlabel('Energy')
ax[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Explore relationship between Track Popularity vs. Danceability and Energy
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# Track Popularity vs. Danceability
sns.scatterplot(data=data, x='danceability', y='track_popularity', ax=ax[0])
ax[0].set_title('Track Popularity vs. Danceability')
ax[0].set_xlabel('Danceability')
ax[0].set_ylabel('Track Popularity')

# Track Popularity vs. Energy
sns.scatterplot(data=data, x='energy', y='track_popularity', ax=ax[1])
ax[1].set_title('Track Popularity vs. Energy')
ax[1].set_xlabel('Energy')
ax[1].set_ylabel('Track Popularity')

plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
from sklearn.preprocessing import StandardScaler

features = [
  'danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
  'instrumentalness', 'liveness', 'valence', 'tempo'
]

selected_data = data[features]

# # standard feature 
scaler = StandardScaler()
features_scaled = scaler.fit_transform(selected_data)

# Dimensionality Reduction for Visualization
pca = PCA(n_components=2)
features_reduced = pca.fit_transform(features_scaled)

# Create a DataFrame for the reduced features and cluster labels
df_reduced = pd.DataFrame(features_reduced, columns=['PC1', 'PC2'])
df_reduced['Cluster'] = data['playlist_genre']

# Plotting the clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_reduced, x='PC1', y='PC2', hue='Cluster', palette='viridis', alpha=0.7)
plt.title('Clusters Visualization with PCA')
plt.show()
```
